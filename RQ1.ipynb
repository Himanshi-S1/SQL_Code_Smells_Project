## **Part 1: Implicit coloumns**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset1=pd.read_csv("sample_data/Implicit_columns_combined.csv")
dataset1.columns
```

```python
# dataset1.head()
dataset1.describe()
```

```python
dataset1.groupby(dataset1['Category'])[['sumDBAccess']].median()
```

```python
print("MEDIAN: ")
dataset1.groupby(dataset1['Category'])[['smellPerAccess']].median()*100
```

```python
print("MEAN: ")
dataset1.groupby(dataset1['Category'])[['smellPerAccess']].mean()*100
```

```python
dataset1.median()
```

```python
dataset1.groupby(dataset1['Category'])[['sumDBAccess']].median()
```

```python
sns.set(rc={'figure.figsize':(6.7,6.7)})
sns.set(font_scale=1.5)
sns.set_style('white')
sns.set_style('ticks', {'axes.edgecolor': '0',  'xtick.color': '0', 'ytick.color': '0'})
plot= sns.boxplot(x= dataset1['Category'], y=dataset1['smellPerAccess']*100, data=dataset1, color='black',medianprops=dict(color='red'), 
                  boxprops = dict(facecolor='gray',edgecolor='black'))
# plot.set_title("Percentage of the ratio of Implicit column smells to number of queries")
# plot.set(xlabel="Category of projects", ylabel="Percentage of smells to queries")
plot.set_xlabel("Category of projects",fontsize=20)
plot.set_ylabel("Prevalence (%)",fontsize=20)
fig = plot.get_figure()
fig.savefig("implicit_column_app_domain.pdf")
```

## **Part 2: Fear of the unknown**

```python
df=pd.read_csv("sample_data/fear_of_unknown_combined.csv")
df.columns
```

```python
df.head()
df.describe()
```

```python
df.mean()
df.groupby(df['Category'])[['smellPerAccess']].median()*100
```

```python
sns.set(rc={'figure.figsize':(8,6)})
sns.set(font_scale=1.5)
sns.set_style('white')
sns.set_style('ticks', {'axes.edgecolor': '0',  
                        'xtick.color': '0',
                        'ytick.color': '0'})
plot= sns.boxplot(x= df['Category'], y=df['smellPerAccess']*100, data=df, color='gray',medianprops=dict(color='red'), 
                  boxprops = dict(facecolor='gray',edgecolor='black' ))
#plot.set_title("Percentage of the ratio of fear of the unknown smells to number of queries")
#plot.set(xlabel="Category", ylabel="Percentage of smells to queries")
plot.set_xlabel("Category of projects",fontsize=20)
plot.set_ylabel("Prevalence (%)",fontsize=20)
fig = plot.get_figure()
fig.savefig("fear_of_unknown_app_domain.pdf")
```
